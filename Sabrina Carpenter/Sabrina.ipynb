```python
import pandas as pd

# Load the CSV file with semicolon delimiter and correct encoding
sab= pd.read_csv('sabrina.csv', sep=';', encoding='utf-8-sig')

# Display the first few rows to verify
sab
```

```python
#the spotify_stream column was a string so i decided to change it to an integer 
sab['spotify_streams'] = sab['spotify_streams'].astype(str)
sab['spotify_streams'] = sab['spotify_streams'].str.replace('.', '', regex=False).replace('', '0').astype(int)
#i dropped uncessary columns i dont need here
columns_to_drop= [
    "track_id", "track_name", "uri", "track_type", 
    "track_musical_genre", "track_videoclip", 
    "videoclip_views", "spotify_global_peak"  # Optional: drop if avoiding target leakage
]

sab_clean=sab.drop(columns=columns_to_drop)
sab_clean
```

```python
#i checked there amount or types of values are in here so i can convert it to numerals since logistic regression cannot train non numerals
sab_clean['album'].unique()
sab_clean['album_musical_genre'].unique()
```

```python
#this is for the name of the albulm
album_mapping = {
    "Short n' Sweet": 1,
    "emails i can’t send fwd": 2,
    "emails i can’t send": 3,
    "Singular: Act II": 4,
    "Singular: Act I": 5,
    "EVOLution": 6,
    "Eyes Wide Open": 7
}

sab_clean['album_numeric'] = sab_clean['album'].map(album_mapping)
sab_clean['album_musical_genre'].unique()


#this is for the genre 
genre_mapping={
    'Pop':1,
    'Electropop':2,
    'Folk Pop': 3
}

sab_clean['genre_numeric']=sab_clean['album_musical_genre'].map(genre_mapping)
sab_clean['genre_numeric'].unique()
print(sab_clean['genre_numeric'].unique())  # Output: [1, 2, 3, 4, 5, 6, 7]
#sab_clean
```

```python
#by doing this i create new columns the goal will be to drop a bunch of these columns which are the previous albums and genre coiumns
new_cols_to_drop=[
    'album',
    'album_musical_genre'
]

clean_sab2=sab_clean.drop(columns=new_cols_to_drop)
clean_sab2
```
